```python
import csv
import pandas as pd
import os
import errno
```

```python
df = pd.read_csv('nca-sample-naomicorr.tsv', delimiter='\t')
```

```python
valListeRows = df[df['form'].str.contains('REF', regex=False)]
```

```python
valIndices = valListeRows.index.values.tolist()
valIndices
```

```python
valListe = df[df['form'].str.contains('REF', regex=False)]
```

```python
listeTitres = valListe['form'].str.extract(r'\[REF:(.*)\]')
listeTitres
```

```python
try: 
    os.mkdir('tableursParTexte') 
except OSError as exc: 
    if exc.errno != errno.EEXIST: 
        raise 
    pass
```

```python
## Ça fait par extrait

for i in range(len(valIndices)):
    print(i)
    j = i + 1
    valIndDf = valIndices[i]+1
    valIndDf2 = valIndices[j]
    dfPret = df.iloc[valIndDf:valIndDf2]
    print(dfPret)
    titre = listeTitres[0].iloc[i]
    dfPret.to_csv('tableursParTexte/'+titre+'.tsv', sep = '\t', index = False)
```
